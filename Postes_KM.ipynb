```python
import pandas as pd

postes_viejo = pd.read_excel("postes_km_anterior.xlsx")

nuevos_nombres = {
    "ruta": "Ruta",
    "km": "KM",
    "longitude": "Lat",
    "latitude": "Long"
}
postes_viejo.rename(columns=nuevos_nombres, inplace=True)

postes_viejo = postes_viejo[["Ruta","KM","Lat","Long"]]
postes_viejo
```

```python
postes_relevados = pd.read_excel("radares_geolocalizacion.xlsx", sheet_name="Postes")
postes_relevados = postes_relevados[["Ruta","KM","Lat","Long"]]
postes_relevados
```

```python
no_coincidencias = []

for _, fila1 in postes_viejo.iterrows():
    for _, fila2 in postes_relevados.iterrows():
        if fila1["Ruta"] == fila2["Ruta"] and fila1["KM"] == fila2["KM"]:
            no_coincidencias.append(fila2)  


df_no_coincidencias = pd.DataFrame(no_coincidencias)

df_no_coincidencias
```
